```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../data/processed/bus_delay_dataset.csv")
print("Dataset shape:", df.shape)
df.head()
```

```python
df.info()
df.describe(include="all")
```

```python
print("Unique trip_id count:", df['trip_id'].nunique())

df['trip_id'].value_counts().head()
```

```python
trip_ids = df['trip_id'].unique()

train_ids, temp_ids = train_test_split(trip_ids, test_size=0.2, random_state=42)

val_ids, test_ids = train_test_split(temp_ids, test_size=0.5, random_state=42)

print("Train IDs:", len(train_ids))
print("Validation IDs:", len(val_ids))
print("Test IDs:", len(test_ids))
```

```python
train_df = df[df['trip_id'].isin(train_ids)]
val_df   = df[df['trip_id'].isin(val_ids)]
test_df  = df[df['trip_id'].isin(test_ids)]

print("Train shape:", train_df.shape)
print("Validation shape:", val_df.shape)
print("Test shape:", test_df.shape)
```

```python
train_df.to_csv("../data/processed/train.csv", index=False)
val_df.to_csv("../data/processed/val.csv", index=False)
test_df.to_csv("../data/processed/test.csv", index=False)

print("Datasets saved in ../data/processed/ as train.csv, val.csv, test.csv")
```
